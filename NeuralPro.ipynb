```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, Input
from tensorflow.keras.optimizers import Adam

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
# =========================================================
# EXPERIMENT 1 - DEEP RELU NETWORK
model_relu = Sequential([
    Input(shape=(28, 28)),
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.3),
    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model_relu.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history_relu = model_relu.fit(
    x_train, y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2
)

relu_loss, relu_accuracy = model_relu.evaluate(x_test, y_test)
print("ReLU Accuracy:", relu_accuracy)
```

```python
# EXPERIMENT 2 - SIGMOID NETWORK
model_sigmoid = Sequential([
    Input(shape=(28, 28)),
    Flatten(),
    Dense(128, activation='sigmoid'),
    Dropout(0.4),
    Dense(32, activation='sigmoid'),
    Dense(10, activation='softmax')
])

model_sigmoid.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history_sigmoid = model_sigmoid.fit(
    x_train, y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.2
)

sigmoid_loss, sigmoid_accuracy = model_sigmoid.evaluate(x_test, y_test)
print("Sigmoid Accuracy:", sigmoid_accuracy)
```

```python
# EXPERIMENT 3 - TANH + LOW LEARNING RATE
model_tanh = Sequential([
    Input(shape=(28, 28)),
    Flatten(),
    Dense(128, activation='tanh'),
    Dropout(0.2),
    Dense(64, activation='tanh'),
    Dense(10, activation='softmax')
])

model_tanh.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history_tanh = model_tanh.fit(
    x_train, y_train,
    epochs=15,
    batch_size=32,
    validation_split=0.2
)

tanh_loss, tanh_accuracy = model_tanh.evaluate(x_test, y_test)
print("Tanh Accuracy:", tanh_accuracy)
```

```python

plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(history_relu.history['val_accuracy'], label='ReLU Val Acc', linestyle='-', marker='o')
plt.plot(history_sigmoid.history['val_accuracy'], label='Sigmoid Val Acc', linestyle='--', marker='s')
plt.plot(history_tanh.history['val_accuracy'], label='Tanh Val Acc', linestyle='-.', marker='^')
plt.title('Validation Accuracy Comparison')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history_relu.history['val_loss'], label='ReLU Val Loss', linestyle='-', marker='o')
plt.plot(history_sigmoid.history['val_loss'], label='Sigmoid Val Loss', linestyle='--', marker='s')
plt.plot(history_tanh.history['val_loss'], label='Tanh Val Loss', linestyle='-.', marker='^')
plt.title('Validation Loss Comparison')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

print("| Experiment | Accuracy | Loss |")
print(f"|  ReLU  | {relu_accuracy:.4f} | {relu_loss:.4f} |")
print(f"| Sigmoid    | {sigmoid_accuracy:.4f} | {sigmoid_loss:.4f} |")
print(f"| Tanh       | {tanh_accuracy:.4f} | {tanh_loss:.4f} |")
```
